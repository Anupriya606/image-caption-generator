```python
import zipfile

with zipfile.ZipFile('/content/flickr8k.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/flickr8k')

print("Unzipped successfully!")
```

```python
import os

# Check folders
print("Folders inside flickr8k:")
print(os.listdir('/content/flickr8k'))

# Count images
image_path = '/content/flickr8k/Images'
images = os.listdir(image_path)
print(f"\nTotal images found: {len(images)}")
print(f"Sample image names: {images[:5]}")
```

```python
import pandas as pd

captions_path = '/content/flickr8k/captions.txt'
df = pd.read_csv(captions_path)

print(f"Total captions: {len(df)}")
print(f"\nColumn names: {df.columns.tolist()}")
print(f"\nFirst 10 rows:")
df.head(10)
```

```python
# Display a sample imgage with its captions:
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Pick a sample image
sample_image = images[0]
sample_captions = df[df['image'] == sample_image]['caption'].tolist()

# Show it
img = mpimg.imread(f'{image_path}/{sample_image}')
plt.figure(figsize=(8, 6))
plt.imshow(img)
plt.axis('off')
plt.title("Sample Image from Flickr8k", fontsize=14)
plt.show()

print("\nCaptions for this image:")
for i, cap in enumerate(sample_captions):
    print(f"{i+1}. {cap}")
```
